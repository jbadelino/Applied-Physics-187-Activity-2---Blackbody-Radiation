```python
#Emittance spectrum of a blackbody radiator 

import numpy as np 
import matplotlib.pyplot as plt
from scipy.constants import c,h,k 

N = 100000

def convert(x): #converts from nanometer to meter
    return x*(1e-9)

#Planck's blackbody radiation formula 
def B(wavelength, T): 
    return (2*h*(c**2))/(convert(wavelength)**5*(np.exp(h*c/(convert(wavelength)*k*T))-1))

#wavelength range 
wavelength = np.linspace(350,750,N) #in nanometers #visible spectrum

#temperature
T = [1000,2500,5400,6500]


#not normalized 
for i in T:
    plt.plot(wavelength, B(wavelength, i), label = "T=%.0f K" % i)
plt.xlabel('wavelength (nm)')
plt.ylabel('emittance')
plt.xlim(350,750)
plt.legend()
plt.show()


#normalized #dividing the curve by the area under the curve
for i in T:
    area = np.trapz(B(wavelength,i), wavelength)
    normalized = B(wavelength,i)/area
    plt.plot(wavelength,normalized, label = "T=%.0f K" % i)
plt.xlabel('wavelength (nm)')
plt.ylabel('emittance')
plt.xlim(350,750)
plt.legend()
plt.show()


#normalize by dividing the maximum value of B, so that B ranges from 0 to 1
for i in T:
    normalized = B(wavelength,i)/max(B(wavelength,i))
    plt.plot(wavelength,normalized, label = "T=%.0f K" % i)
plt.xlabel('wavelength (nm)')
plt.ylabel('emittance')
plt.xlim(350,750)
plt.legend()
plt.show()


```
